```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
# Given data
n1 = 10
mean1 = 9
std1 = 2

n2 = 10
mean2 = 8.5
std2 = 1.5

# Calculate sample variances
variance1 = std1**2
variance2 = std2**2

# Compute F-statistic
F_statistic = variance1 / variance2

# Degrees of freedom
df1 = n1 - 1
df2 = n2 - 1

# Calculate p-value
p_value = scipy.stats.f.sf(F_statistic, df1, df2)

# Significance level
alpha = 0.05

# Print results
print("F-statistic:", F_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject null hypothesis: Variances are not significantly different.")
```

## Question 3

```python
# Given data
n1 = 8
std1 = 10

n2 = 6
std2 = 7

# Check which variance is larger
if std1 > std2:
    variance1, df1 = std1**2, n1 - 1
    variance2, df2 = std2**2, n2 - 1
else:
    variance1, df1 = std2**2, n2 - 1
    variance2, df2 = std1**2, n1 - 1

# Compute F-statistic
F_statistic = variance1 / variance2

# Calculate p-value
p_value = scipy.stats.f.sf(F_statistic, df1, df2)

# Significance level
alpha = 0.02

# Calculate critical value
critical_f_value = scipy.stats.f.ppf(1 - alpha / 2, df1, df2)

# Print results
print("F-statistic:", F_statistic)
print("p-value:", p_value)
print("Critical value:", critical_f_value)

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject null hypothesis: Variances are not significantly different.")

# Plotting
x = np.linspace(0, 15, 1000)
y = scipy.stats.f.pdf(x, df1, df2)
plt.plot(x, y, color='black', lw=2, label='F-distribution')
plt.fill_between(x[x > critical_f_value], 0, scipy.stats.f.pdf(x[x > critical_f_value], df1, df2),
                 color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x[x > F_statistic], 0, scipy.stats.f.pdf(x[x > F_statistic], df1, df2),
                 color='darkgreen', alpha=0.3, label='Area beyond F-statistic')
plt.axvline(x=F_statistic,color='green',linestyle='--',label='F-statistic')
plt.axvline(x=critical_f_value,color='red',linestyle='--',label='Critical Value')
plt.xlabel('F-value')
plt.ylabel('Probability Density')
plt.title('F-distribution with Rejection Region')
plt.legend()
plt.grid(True)
plt.show()
```

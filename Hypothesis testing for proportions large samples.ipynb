```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python

def calculate_z_score(sample_proportion, population_proportion, sample_size):
    """Calculate the z-score for hypothesis testing."""
    return (sample_proportion - population_proportion) / ((population_proportion*(1-population_proportion))/sample_size)**0.5

def perform_hypothesis_test(z_score, alpha,tail):
    """Perform the hypothesis test and plot the normal distribution."""
    # Calculate critical value (one-tailed test)
    if tail.lower() == 'upper':
        critical_value = stats.norm.ppf(1 - alpha)
    elif tail.lower() == 'lower':
        critical_value = stats.norm.ppf(alpha)

    # Print the z-score and critical value
    print("Z-score:", z_score)
    print("Critical Value:", critical_value)

    # Perform hypothesis test
    if np.abs(z_score) > np.abs(critical_value):
        print("Reject Null Hypothesis (H0)")
    else:
        print("Fail to Reject Null Hypothesis (H0)")

    # Plot the normal distribution
    x = np.linspace(-4, 4, 100)
    y = stats.norm.pdf(x, 0, 1)
    plt.plot(x, y)
    if tail.lower() == 'upper':
        plt.fill_between(x[x > critical_value], 0, stats.norm.pdf(x[x > critical_value], 0, 1), color='red', alpha=0.5)
    elif tail.lower() == 'lower':
        plt.fill_between(x[x < critical_value], 0, stats.norm.pdf(x[x < critical_value], 0, 1), color='red', alpha=0.5)
    plt.axvline(x=critical_value,linestyle='--',color='black')
    plt.axvline(x=z_score,linestyle='--',color='green')
    plt.text(critical_value + 0.1, 0.1, 'Critical Region', fontsize=10, color='red')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.title('Normal Distribution with Critical Region')
    plt.show()
```

```python
def plot_two_tailed_hypothesis_test(alpha,z):
    # Generate x values for the plot
    x = np.linspace(-4, 4, 1000)

    # Calculate the critical values for a two-tailed test
    critical_value_left = stats.norm.ppf(alpha / 2)
    critical_value_right = stats.norm.ppf(1 - alpha / 2)
    print(f'Critical Values {(critical_value_left,critical_value_right)}')
    print(f'Z-score:{z}')
    # Plot the standard normal distribution
    plt.plot(x, stats.norm.pdf(x, 0, 1), label='Standard Normal Distribution')

    # Shade the left critical region
    plt.fill_between(x[x < critical_value_left], 0, stats.norm.pdf(x[x < critical_value_left], 0, 1), color='red', alpha=0.5)

    # Shade the right critical region
    plt.fill_between(x[x > critical_value_right], 0, stats.norm.pdf(x[x > critical_value_right], 0, 1), color='red', alpha=0.5)

    # Plot vertical lines at the critical values
    plt.axvline(critical_value_left, color='red', linestyle='--', label='Critical Values')
    plt.axvline(critical_value_right, color='red', linestyle='--')
    plt.axvline(z, color='green', linestyle='--')


    # Add labels and legend
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title('Two-Tailed Hypothesis Test')
    plt.legend()

    # Show plot
    plt.show()
```
